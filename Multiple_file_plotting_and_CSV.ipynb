```python
import re
import numpy as np
import pandas as pd
import os
import glob
import sounddevice as sd
from matplotlib import pyplot as plt
from datetime import datetime
from scipy.io.wavfile import write
```

# Saving all files in a folder to CSV

```python
give_input_folder_path_here=str(os.getcwd())+'\\DATA'
give_output_CSV_folder_path_here=str(os.getcwd())+'\\DATA'
give_output_Wav_folder_path_here=str(os.getcwd())+'\\DATA'
date_format = '%Y-%m-%d %H:%M:%S.%f' 
# Change  date_format according to the value you get because the timestamp is generated by the server and may vary for every server
```

```python
for filename in glob.glob(os.path.join(give_input_folder_path_here, '*.txt')):
    with open(os.path.join(os.getcwd(), filename), 'r') as f:
        f_name=filename.split('\\')[-1]
        f_name=f_name.split('.')[0]
        aa=str(filename[:-4])
        string=f.read()
        string=string.replace("\x00","")
        pattern = "\[(.*?)\]"
        timestamps = re.findall(pattern, string)
        totaltime=datetime.strptime(timestamps[-1], date_format)-datetime.strptime(timestamps[0], date_format)
        timeinsec=totaltime.total_seconds()
        pattern = "\((.*?)\)"
        Microphone_temp = re.findall(pattern, string)
        Microphone_temp=[re.sub('\[(.*?)\]', '', x) for x in Microphone_temp]
        Microphone=[x for x in Microphone_temp if len(x)<12]
        Microphone_numpy=np.array(Microphone)
        Microphone_numpy = Microphone_numpy.astype(int)
        Sr=len(Microphone_temp)/timeinsec
        print("Sample Rate of microphone for {} are: {}".format(f_name,Sr))
        sound_df=pd.DataFrame(Microphone_numpy,columns=['Microphone'])
        sound_df.to_csv(give_output_CSV_folder_path_here+"\\"+f_name+'_Microphone.csv',encoding='utf-8')
        scaled = np.int16(Microphone_numpy/np.max(np.abs(Microphone_numpy)) * 32767)
        write(give_output_Wav_folder_path_here+"\\"+f_name+'_Microphone.wav', int(Sr), scaled)
            
```
